```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv("BTCUSD_Candlestick_1_D_BID_07.03.2022-06.03.2025.csv")
df
```

```python
df = df[df["Volume"] != 0]
df.reset_index(drop=True, inplace=True)
df.tail(5)
```

```python
dfpl = df[0:1095]

fig = go.Figure(data=[go.Candlestick(x=dfpl.index,
                                     open=dfpl['Open'],
                                     high=dfpl['High'],
                                     low=dfpl['Low'],
                                     close=dfpl['Close'])])

fig.show()
```

```python
data = yf.download(tickers="BTC-USD",
                   period="max",
                   interval="1d")
```

```python
data.tail(5)
```

```python
data.reset_index(inplace=True)
data.columns = data.columns.droplevel(1)
data.tail(5)
```

```python
dfpl = data[2730:3825]

fig = go.Figure(data=[go.Candlestick(x=dfpl.index,
                                     open=dfpl['Open'],
                                     high=dfpl['High'],
                                     low=dfpl['Low'],
                                     close=dfpl['Close'])])

fig.show()
```

```python
import pandas_ta as ta
df['EMA'] = ta.ema(df.Close, length=10)
df['RSI_10'] = ta.rsi(df.Close, length=10)
df.tail(5)
```

```python
dfpl = df[10:1095]

fig = go.Figure(data=[go.Candlestick(x=dfpl.index,
                                     open=dfpl['Open'],
                                     high=dfpl['High'],
                                     low=dfpl['Low'],
                                     close=dfpl['Close']),
                go.Scatter(x=dfpl.index, y=dfpl.EMA, line=dict(color='red', width=2), name='EMA')])
fig.show()
```
