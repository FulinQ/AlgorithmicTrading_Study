```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv("XAGUSD_Candlestick_4_Hour_BID_01.03.2023-01.03.2025.csv")
df
```

```python
df = df[df["Volume"] != 0]
df.reset_index(drop=True, inplace=True)
df.tail(5)
```

```python
dfpl = df[3000:3204]

fig = go.Figure(data=[go.Candlestick(x=dfpl.index,
                                     open=dfpl['Open'],
                                     high=dfpl['High'],
                                     low=dfpl['Low'],
                                     close=dfpl['Close'])])

fig.show()
```
